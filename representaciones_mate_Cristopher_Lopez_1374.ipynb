Problema 1: La suma de un número más 9 es igual a 21Paso 1: Explicación y ResoluciónPlanteamiento: Identificamos la incógnita. "Un número" lo representaremos con la variable $x$.La Ecuación: La frase "la suma de un número más 9 es igual a 21" se traduce como:$$x + 9 = 21$$Despeje: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Resultado:$$x = 12$$

```python
# Problema 1: Encontrar el número desconocido
total = 21
numero_conocido = 9

# Despeje: x = total - numero_conocido
x = total - numero_conocido

print(f"El número buscado es: {x}")
```

### Problema 2:
 #### Las edades de Pedro y AnaPaso 1: Explicación y ResoluciónPlanteamiento:Edad de Ana: $x$ (es nuestra referencia).Edad de Pedro: $2x$ (porque es el doble que la de Ana).La Ecuación: "Entre los dos suman 36 años":$$x + 2x = 36$$
 ### Simplificación:
  ####Sumamos las variables semejantes:$$3x = 36$$
  ### Despeje:
  #### El 3 que multiplica a la $x$ pasa dividiendo al 36:$$x = \frac{36}{3}$$
  ### Resultado:
  #### Ana ($x$): 12 años.Pedro ($2x$): $2 \times 12 = 24$ años.
  ### Comprobación:
   $12 + 24 = 36$.

```python
# Problema 2: Edades de Pedro y Ana
suma_total = 36

# La ecuación es x + 2x = 36 -> 3x = 36
# Por lo tanto, x = suma_total / 3
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
