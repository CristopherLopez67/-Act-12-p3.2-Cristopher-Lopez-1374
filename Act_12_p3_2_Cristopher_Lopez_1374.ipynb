```python
# Pida al usuario ingresar un número y lo guarde en una variable llamada NÚM.
# Use una estructura if-else para evaluar si NÚM >= 0.
# Si es verdadero, asigna el texto 'POSITIVO' a una variable R.
# Si es falso, asigna el texto 'NEGATIVO' a la variable R.
# Finalmente, imprime el valor de R.
num = float(input("Ingrese un número: "))
if num >= 0:
    r = "POSITIVO"
else:
    r = "NEGATIVO"
print(r)

print("Fin del programa")
print("origrama realizado por cristopher lopez NL = 30")
```
